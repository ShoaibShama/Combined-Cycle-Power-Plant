```python
import numpy as np
import pandas as pd
from sklearn import preprocessing 
```

```python
data_train=np.loadtxt('training_ccpp_x_y_train.csv',delimiter=',')
df=pd.DataFrame(data_train[:,:-1])
df.columns=["T","V","AP","RH"]
df['T_RH']=df['T']*df['RH']
df['T_V']=df['T']*df['V']
X=df.values
Y=data_train[:,-1]
scaling=preprocessing.StandardScaler()
scaling.fit(X)
X=scaling.transform(X)
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(X,Y)
```

```python
def score(y_true,y_predict):
    u=((y_true-y_predict)**2).sum()
    v=((y_true-y_true.mean())**2).sum()
    return 1-(u/v)

def step_gradient(x_train,y_train,learning_rate,m):
    m_slope=np.zeros(x_train.shape[1]+1,dtype=float)
    N=len(x_train)
#     x_train = train_data[:,:-1]
#     y_train = train_data[:, -1]
    for i in range(N):
        x=x_train[i]
        x=np.append(x,1)
        for j in range(x_train.shape[1]+1):
            m_slope[j]+=(-2/N)*(y_train[i]-(x*m).sum())*x[j]
    new_m= m-( learning_rate*m_slope)
    return new_m;
def cost(x_train,y_train,m):
    total_cost=0
#     x_train = train_data[:,:-1]
#     y_train = train_data[:, -1]
    for i in range(len(x_train)):
        x=x_train[i]
        x=np.append(x,1)
        total_cost+= (1/len(x_train))*((y_train[i]-sum(m * x))**2)
    return total_cost
def gradient_decent(x_train,y_train,number_of_itration,learning_rate):
    m=np.zeros(x_train.shape[1]+1,dtype=float)
    N=len(x_train)
    for i in range(number_of_itration):
        m=step_gradient(x_train,y_train,learning_rate,m)
        print("Cost :",cost(x_train,y_train,m))
    return m
def gd(x_train,y_train):
    learning_rate=0.01
    number_of_itration=450
    m=gradient_decent(x_train,y_train,number_of_itration,learning_rate)
    return m
```

```python
m=gd(x_train,y_train)
m
```

```python
prediction=[]
for i in x_test:
    prediction.append((i*m[:-1]).sum()+m[-1])
prediction
print("Score :",score(y_test,prediction))
```
